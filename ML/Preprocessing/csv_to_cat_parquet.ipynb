```python
import pandas as pd
import numpy as np

from sklearn.model_selection import StratifiedKFold
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
pd.set_option('display.max_colwidth',None)
```

```python
df = pd.read_csv('original_data/train_dataset_Самолет.csv')
df.columns = df.columns.str.lower()
df.head()
```

```python
cat_columns = [
    'col520', 'col552', 'col543', 'col527', 'col607', 'col600', 'col599',
    'col560', 'col548', 'col551', 'col535', 'col540', 'col464', 'col152', 
    'col228', 'col559', 'col460', 'col472', 'col568', 'col468'
]

df[cat_columns] = df[cat_columns].astype('str')

for col_name in cat_columns:
    df.loc[df[col_name]=='nan', col_name] = 'miss_value'
    mask = df[col_name].value_counts(normalize=True)
    df.loc[df[col_name].isin(list(mask[mask<0.0005].index)), col_name] = 'other'
```

```python
df[cat_columns].head()
```

```python
bad_cat = list(df.select_dtypes('object').columns)
cat_columns = cat_columns + ['report_date']
bad_cat = [f for f in bad_cat if f not in cat_columns]
df = df.drop(bad_cat, axis=1)
```
